```python
file = open("userlog.log")
data = file.readlines()

date=[]
user=[]
for i in range(5000):
    each_line_list = data[i].split()
    each_line_list_date = each_line_list[0]
    each_line_list_user = each_line_list[4]
    date.append(each_line_list_date)
    user.append(each_line_list_user)

date_set=set(date)
user_set=set(user)
date_list=list(date_set)
user_list=list(user_set)
date_list_sort=['2020-05-01', '2020-05-02', '2020-05-03', '2020-05-04', '2020-05-05', '2020-05-06', '2020-05-07', '2020-05-08', '2020-05-09', '2020-05-10', '2020-05-11', '2020-05-12', '2020-05-13', '2020-05-14', '2020-05-15', '2020-05-16', '2020-05-17', '2020-05-18', '2020-05-19', '2020-05-20', '2020-05-21', '2020-05-22', '2020-05-23', '2020-05-24', '2020-05-25', '2020-05-26', '2020-05-27', '2020-05-28', '2020-05-29', '2020-05-30']
```

```python
result="=========================\n==="+ " DailyActivityCount "+ " ===\n=========================\n"
for j in date_list_sort:
    login_count=0
    logout_count=0
    for k in range(5000):
        each_line_list = data[k].split()
        each_line_list_activity = each_line_list[2]
        each_line_list_date = each_line_list[0]
        if j==each_line_list_date and each_line_list_activity=='login':
            login_count+=1
        if j==each_line_list_date and each_line_list_activity=='logout':
            logout_count+=1
    result+=j+"\t login count: "+str(login_count)+"\t logout count: " + str(logout_count)+"\n"
#print(result)

file2 = open("dailyactivitycount_report.txt", "w")
file2.write(result)
file2.close()
```

```python
result2="=========================\n==="+ " DailyServerCount "+ " ===\n=========================\n"
result8="=========================\n==="+ " DailyServerMostPopular "+ " ===\n=========================\n"
result9="=========================\n==="+ " DailyServerLeastPopular "+ " ===\n=========================\n"
for j in date_list_sort:
    mailserver=0
    myworkstation=0
    webserver=0
    for k in range(5000):
        each_line_list = data[k].split()
        each_line_list_server = each_line_list[3]
        each_line_list_date = each_line_list[0]
        if j==each_line_list_date and each_line_list_server=='mailserver.local':
            mailserver+=1
        if j==each_line_list_date and each_line_list_server=='myworkstation.local':
            myworkstation+=1
        if j==each_line_list_date and each_line_list_server=='webserver.local':
            webserver+=1
    result2+=j+"\t mailserver.local: "+str(mailserver)+"\t myworkstation.local: " + str(myworkstation)+"\t webserver.local: " + str(webserver)+"\n"
    
    if mailserver>myworkstation and mailserver>webserver:
        result8+=j+"\t mailserver.local"+"\n"
    if mailserver<myworkstation and myworkstation>webserver:
        result8+=j+"\t myworkstation.local"+"\n"
    if mailserver<webserver and myworkstation<webserver:
        result8+=j+"\t webserver.local"+"\n"
    
    if mailserver<myworkstation and mailserver<webserver:
        result9+=j+"\t mailserver.local"+"\n"
    if mailserver>myworkstation and myworkstation<webserver:
        result9+=j+"\t myworkstation.local"+"\n"
    if mailserver>webserver and myworkstation>webserver:
        result9+=j+"\t webserver.local"+"\n"


file3 = open("dailyservercount_report.txt", "w")
file3.write(result2)
file3.close()

file8 = open("dailyservermostpopular_report.txt", "w")
file8.write(result8)
file8.close()

file9 = open("dailyserverleastpopular_report.txt", "w")
file9.write(result9)
file9.close()
```

```python
result3="=========================\n==="+ " ServerDistribution "+ " ===\n=========================\n"
mailserver2=0
myworkstation2=0
webserver2=0
total2=0
for m in range(5000):
    each_line_list = data[m].split()
    each_line_list_server = each_line_list[3]
    if each_line_list_server=='mailserver.local':
        mailserver2+=1
        total2+=1
    if each_line_list_server=='myworkstation.local':
        myworkstation2+=1
        total2+=1
    if each_line_list_server=='webserver.local':
        webserver2+=1
        total2+=1
result3+=" mailserver.local: "+str(mailserver2)+"\n myworkstation.local: " + str(myworkstation2)+"\n webserver.local: " + str(webserver2)+"\n total: "+str(total2)+"\n"
#print(result3)

file4 = open("serverdistribution_report.txt", "w")
file4.write(result3)
file4.close()
```

```python
result4="=========================\n==="+ " UserActivityCount "+ " ===\n=========================\n"
for each_user in user_list:
    each_user_count=0
    for k in range(5000):
        each_line_list = data[k].split()
        each_line_list_user = each_line_list[4]
        if each_user==each_line_list_user:
            each_user_count+=1
    result4+=each_user+"\t "+str(each_user_count) +"\n"
#print(result4)

file5 = open("useractivitycount_report.txt", "w")
file5.write(result4)
file5.close()
```

```python
letters_to_names = {}
for each_user2 in user_list:
    letter = each_user2[0]
    namelist= letters_to_names.get(letter, [])
    namelist.append(each_user2)
    letters_to_names.update({letter: namelist})

#print(letters_to_names)
result5="=========================\n==="+ " UserNameStartLetter "+ " ===\n=========================\n"
for keys in letters_to_names:
    result5+=keys+str(letters_to_names[keys])+"\n"
#print(result5)

file6 = open("usernamestartletter_report.txt", "w")
file6.write(result5)
file6.close()
```
